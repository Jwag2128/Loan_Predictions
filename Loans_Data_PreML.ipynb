```python
# Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
import statistics
```

```python
# Import CSV file
loans_prepro_df = pd.read_csv('Loans_Data_Clean.csv')
loans_prepro_df
```

```python
# Change data type of columns needed for sum column
loans_prepro_df.astype({'open_il_6m': 'int64', 'open_il_12m': 'int64', 'open_il_24m': 'int64'}).dtypes
loans_prepro_df['Open_Acc_IL24m'] = loans_prepro_df['open_il_6m'] + loans_prepro_df['open_il_12m'] + loans_prepro_df['open_il_24m']

print(loans_prepro_df['Open_Acc_IL24m'])
```

```python
# Drop unused combined columns
loans_prepro_df = loans_prepro_df.drop(columns=['open_il_6m', 'open_il_12m', 'open_il_24m'], axis=0)
loans_prepro_df.head()
```

```python
# Change data type of columns needed for sum revolving account column
loans_prepro_df.astype({'open_rv_12m': 'int64', 'open_rv_24m': 'int64'}).dtypes
loans_prepro_df['Open_RevAcc_24m'] = loans_prepro_df['open_rv_12m'] + loans_prepro_df['open_rv_24m']
loans_prepro_df = loans_prepro_df.drop(columns=['open_rv_12m', 'open_rv_24m'], axis=0)
print(loans_prepro_df['Open_RevAcc_24m'])
loans_prepro_df.head()
```

```python
# Change data type of columns needed for sum past due account column
loans_prepro_df.astype({'num_tl_120dpd_2m': 'int64', 'num_tl_30dpd': 'int64', 'num_tl_90g_dpd_24m': 'int64'}).dtypes
loans_prepro_df['Acc_Curr_PD30days'] = loans_prepro_df['num_tl_120dpd_2m'] + loans_prepro_df['num_tl_30dpd'] + loans_prepro_df['num_tl_90g_dpd_24m']
loans_prepro_df = loans_prepro_df.drop(columns=['num_tl_120dpd_2m', 'num_tl_30dpd', 'num_tl_90g_dpd_24m'], axis=0)
print(loans_prepro_df['Acc_Curr_PD30days'])
loans_prepro_df.head()
```

```python
# Rename Open acc column
loans_prepro_df = loans_prepro_df.rename(columns={'open_acc': '#_Open_Accounts'})
loans_prepro_df.astype({'#_Open_Accounts': 'int64'}).dtypes
loans_prepro_df.head()
```

```python
# Rename Bankruptcies column
loans_prepro_df = loans_prepro_df.rename(columns={'pub_rec_bankruptcies': '#_Bankruptcies'})
loans_prepro_df.astype({'#_Bankruptcies': 'int64'}).dtypes
loans_prepro_df.head()
```

```python
# Create new ML dataframe 
loans_ml_df = loans_prepro_df.filter(items=['Loan_Amt', 'Loan_Term', 'Monthly_Payment', 'Credit_Grade', 'Employement_Length',
                                           'Annual_Income', 'DTI_Ratio', 'Application_Type', 'Joint_Annual_Income',
                                           'Joint_DTI_Ratio', '#_Open_Accounts', '#_Delinquent_Accounts', 'Total_Amount_in_Collection',
                                           'Total_Current_Balance_All', 'Average_Current_Balance', 'Chargeoff_within_Last_Year',
                                           'Delinquent_Amount', '#_Mortgage_Accounts', '#_Bankruptcies', '#_Tax_Liens',
                                           'Balance_Exc_Mortgage', 'Joint_Revolving_Balance', 'Open_Acc_IL24m',
                                           'Open_RevAcc_24m', 'Acc_Curr_PD30days', 'Loan_Status'])
```

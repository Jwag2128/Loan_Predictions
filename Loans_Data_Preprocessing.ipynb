```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Load in datafile
loans_df = pd.read_csv('LoanStats_2017Q1 2.csv')
loans_df
```

```python
# Drop duplicate ID column
loans_df = loans_df.drop(columns="id", axis=0)
loans_df
```

```python
loans_df['member_id'].dtypes
```

```python
# Change NaN in member_id to zero and cast as interger
loans_df['member_id'] = loans_df['member_id'].fillna(0)
loans_df.astype({'member_id': 'int64'}).dtypes
loans_df
```

```python
# Assign member_id values
loans_df['member_id'] = np.arange(1, 96780)
loans_df
```

```python
# remove columns with duplicate information
loans_df = loans_df.drop(columns=["funded_amnt", "funded_amnt_inv", "title"], axis=0)
loans_df
```

```python
# Remove columns with unnecessary data
loans_df = loans_df.drop(columns=["grade", "issue_d", "pymnt_plan", "url", "desc", "initial_list_status",
                                  "recoveries", "collection_recovery_fee", "policy_code"], axis=0)
loans_df
```

```python
# Remove columns with incomplete info or info not needed
loans_df = loans_df.drop(columns=["zip_code", "earliest_cr_line", "out_prncp_inv", "total_pymnt_inv",
                                 "last_pymnt_d", "last_pymnt_d", "next_pymnt_d", "last_credit_pull_d"], axis=0)
loans_df
```

```python
# Remove the word 'month' from term column and change datatype to int
loans_df['term'] = loans_df['term'].str.replace('months', '')
loans_df.astype({'term': 'int64'}).dtypes
loans_df
```

```python
# Replace NaN value in emp_title column
loans_df['emp_title'] = loans_df['emp_title'].fillna('None')
loans_df
```

```python
# Change term column datatype
loans_df.astype({'term': 'int64'}).dtypes
```

```python
# Replace NaN value in emp_length column
loans_df['emp_length'] = loans_df['emp_length'].fillna('None')
loans_df
```

```python
# Replace NaN value in verification_status_joint column
loans_df['verification_status_joint'] = loans_df['verification_status_joint'].fillna('Not Applicable')
loans_df
```

```python
# Replace all other NaN values with 0 
loans_clean_df = loans_df.fillna(0)
loans_clean_df
```

```python
# export to prelim csv file
loans_clean_df.to_csv('prelim_loans_data.csv', index=False)
```

```python
# Rename some columns
loans_data_clean_df = loans_clean_df.rename(columns={'member_id': 'Applicant_ID',
                                                    'loan_amnt': 'Loan_Amt',
                                                    'term': 'Loan_Term',
                                                    'int_rate': 'Interest_Rate',
                                                    'installment': 'Monthly_Payment',
                                                    'sub_grade': 'Credit_Grade',
                                                    'emp_title': 'Job_Title',
                                                    'emp_length': 'Employement_Length',
                                                    'home_ownership': 'Rent_Own',
                                                    'annual_inc': 'Annual_Income',
                                                    'verification_status': 'Income_Verification',
                                                    'loan_status': 'Loan_Status',
                                                    'purpose': 'Loan_Purpose',
                                                    'addr_state': 'State_of_Residence',
                                                    'dti': 'DTI_Ratio',
                                                    'revol_bal': 'Revolving_Balance',
                                                    'out_prncp': 'Outstanding_Principal',
                                                    'application_type': 'Application_Type',
                                                    'annual_inc_joint': 'Joint_Annual_Income',
                                                    'dti_joint': 'Joint_DTI_Ratio',
                                                    'verification_status_joint': 'Joint_Income_Verification',
                                                    'acc_now_delinq': '#_Delinquent_Accounts',
                                                    'tot_coll_amt': 'Total_Amount_in_Collection',
                                                    'tot_cur_bal': 'Total_Current_Balance_All',
                                                    'avg_cur_bal': 'Average_Current_Balance',
                                                    'chargeoff_within_12_mths': 'Chargeoff_within_Last_Year',
                                                    'delinq_amnt': 'Delinquent_Amount',
                                                    'mort_acc': '#_Mortgage_Accounts',
                                                    'tax_liens': '#_Tax_Liens',
                                                    'total_bal_ex_mort': 'Balance_Exc_Mortgage',
                                                    'revol_bal_joint': 'Joint_Revolving_Balance'})
loans_data_clean_df
```

```python
# Cleaned Data Df to CSV file
loans_data_clean_df.to_csv('Loans_Data_Clean.csv', index=False)
```
